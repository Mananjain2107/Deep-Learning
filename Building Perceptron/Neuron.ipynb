# Building a Perceptron

```python
# The original perceptron uses a step function as its activation, but modern neural networks avoid it because it is not differentiable
# So we are also not implementing the step activation function
```

```python
import torch # This imports the PyTorch library, PyTorch is a deep learning framework used to build and train neural networks
import torch.nn as nn # torch.nn contains everything needed to build neural networks like layers, loss functions ,activation functions
```

### Input

```python
torch.manual_seed(42) # fixes randomness so results are consistent across runs
# It is used to make results reproducible (same every time you run the code)
```

```python
inputs = torch.tensor([1.0,2.0,3.0]) # Function used to create a tensor in PyTorch or a function that converts data into a PyTorch tensor
# In deep learning, data is usually converted to float because neural networks require precise numerical computations
```

```python
# As we know a perceptron is a single neuron model that takes multiple inputs and produces one output

neuron = nn.Linear(in_features=3, out_features=1) # nn.Linear() computes the weighted sum of inputs and adds a bias
# Define how many inputs go in and how many outputs (neurons) come out
```

### Output

```python
output = neuron(inputs)
```

```python
output
```

### Verifying Our Output(by applying formula)

```python
# We know that in perceptron we multiply inputs by weights → add bias → get output
neuron.weight @ inputs + neuron.bias # @ performs a dot product
```
